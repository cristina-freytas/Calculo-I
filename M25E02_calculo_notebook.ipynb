```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

# Ignorar avisos
warnings.filterwarnings("ignore")

# Fun√ß√£o para o cubo de x
f = lambda x: x**3

# Fun√ß√£o para calcular a soma de Riemann
def calcula_area(a, b, func, frame_num):
    num_bins = frame_num + 1  # Quantidade de ret√¢ngulos
    delta = (b - a) / num_bins  # Largura do ret√¢ngulo
    x_cols = np.linspace(a, b - delta, num_bins) + delta / 2  # Divis√£o do eixo X
    y_cols = func(x_cols)  # Alturas dos ret√¢ngulos
    return sum(delta * y_cols)

# Fun√ß√£o para anima√ß√£o (desenhando as barras e o gr√°fico)
def animate(n, a=1, b=2, func=f, verdadeiro_valor=4.5):  # Intervalo ajustado para [1, 2]
    plt.cla()  # Limpa o gr√°fico anterior para montar o pr√≥ximo frame
    ax = plt.axes(xlim=(a, b), ylim=(0, 9))  # Monta o sistema de eixos

    # Plota a fun√ß√£o
    x = np.linspace(a, b, 100)
    y = func(x)
    ax.plot(x, y, '-')

    # Define as quebras e as alturas das colunas
    num_bins = n + 1
    delta = (b - a) / num_bins
    x_cols = np.linspace(a, b - delta, num_bins) + delta / 2
    y_cols = func(x_cols)

    # Plota as barras
    bar = ax.bar(x_cols, y_cols, alpha=0.5, width=delta, edgecolor='blue')

    # √Årea dos ret√¢ngulos
    soma_riemann = calcula_area(a, b, func, n)

    # Erro
    erro = verdadeiro_valor - soma_riemann
    ax.set_title(f'Soma de Riemann = {soma_riemann:.2f} erro = {erro:.4f}')

    return bar

# Chama a fun√ß√£o animate com o n√∫mero de divis√µes (n)
animate(50)
```

```python
%matplotlib inline



fig = plt.figure(figsize=(8, 6))  # make a new figure
ax = plt.subplot(1,1,1)   

ani = FuncAnimation(fig, animate, frames=40, interval=120, blit=True)
# ani.save('riemann_sum2.gif',writer="imagemagick")
HTML(ani.to_jshtml())
```
